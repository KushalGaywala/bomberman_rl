```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('2-normal-10000.csv')

# Set up the plot style
plt.style.use('default')
sns.set_palette("deep")

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the average loss
ax.plot(df['Round'], df['Avg_Loss'], label='Average Loss', linewidth=2)

# Set labels and title
ax.set_xlabel('Round', fontsize=12)
ax.set_ylabel('Average Loss', fontsize=12)
ax.set_title('Average Loss per Round over 10,000 Rounds', fontsize=14)

# Add legend
ax.legend(fontsize=10)

# Add grid
ax.grid(True, linestyle='--', alpha=0.7)

# Customize x-axis ticks
ax.set_xticks(range(0, 10001, 1000))
ax.set_xticklabels(range(0, 10001, 1000), rotation=45)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()

# Optional: Save the figure
# plt.savefig('average_loss_per_round.png', dpi=300, bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Read the CSV file
df = pd.read_csv('2-coin-heaven-10000.csv')

# Set up the plot style
plt.style.use('default')
sns.set_palette("deep")

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the average loss using a logarithmic scale
ax.semilogy(df['Round'], df['Avg_Loss'], label='Average Loss', linewidth=2)

# Set labels and title
ax.set_xlabel('Round', fontsize=12)
ax.set_ylabel('Average Loss (log scale)', fontsize=12)
ax.set_title('Average Loss per Round over 10,000 Rounds', fontsize=14)

# Add legend
ax.legend(fontsize=10)

# Add grid
ax.grid(True, linestyle='--', alpha=0.7)

# Customize x-axis ticks
ax.set_xticks(range(0, 10001, 1000))
ax.set_xticklabels(range(0, 10001, 1000), rotation=45)

# Customize y-axis ticks
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0e}'.format(y)))

# Set y-axis limits to focus on the relevant range
y_min = max(df['Avg_Loss'].min(), 1e-6)  # Avoid log(0)
y_max = df['Avg_Loss'].max()
ax.set_ylim(y_min, y_max)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
